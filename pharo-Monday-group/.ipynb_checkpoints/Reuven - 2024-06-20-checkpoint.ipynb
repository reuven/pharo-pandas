# Some common `read_csv` options

1. `usecols` -- pass a list of strings (or integers) of the columns you want to include
2. `header` -- this indicates the line of the file (starting at 0) on which the column names are written. I can pass an integer for the line on which the headers sit, or I can pass `None` to indicate that there are no headers
3. `names` -- a list of strings, the names we want to give to the columns. This is especially useful if we say `header=None`

```python
df = pd.read_csv(filename,
                 usecols=['passenger_count', 
                          'trip_distance',
                          'total_amount'])
df.head()
```
